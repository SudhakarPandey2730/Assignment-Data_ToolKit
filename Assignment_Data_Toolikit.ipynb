**Question 1 : Demonstrate three different methods for creating identical 2D arrays in NumPy. Provide the code for each

## Solution:

Three different methods for creating identical 2D arrays in NumPy are:

**Question 3 :  Explain the following terms:**

*   **The difference in np.array, np.asarray and np.asanyarray.**
*   **The difference between Deep copy and shallow copy.**






## Solution:

### The Difference Between np.array(), np.asarray(), and np.asanyarray()

**np.array():**

This function always creates a new NumPy array.

If you pass an existing array, it copies the data to a new array by default (unless copy=False is specified).

It ensures that the returned object is a base NumPy ndarray, even if you pass in a subclass of ndarray.

```python
import numpy as np
a = [1, 2, 3]
arr1 = np.array(a)
print(type(arr1))
```

**np.asarray():**

This function converts the input to a NumPy array without copying if the input is already a NumPy array (or if the data type and order are already correct).

It is useful when you want to avoid unnecessary copying for performance reasons.

Unlike np.array(), it will not copy the data if the input is already an ndarray.

```python
arr2 = np.asarray(arr1)
print(arr1 is arr2)  # Output: True (no copy if input is already an ndarray)
```

**np.asanyarray():**

Similar to np.asarray(), but it allows the output to be any subclass of ndarray.

If you pass a subclass (e.g., a matrix or masked array), it will return that subclass rather than a base ndarray.

```python
arr3 = np.asanyarray(arr1)
print(type(arr3))  # Output: <class 'numpy.ndarray'>
```

Conclusion:

np.array(): Always returns a new ndarray (or its copy).

np.asarray(): Returns an ndarray, but does not copy if the input is already an array.

np.asanyarray(): Like np.asarray(), but returns any array subclass (e.g., matrix, masked_array) without converting to ndarray.

### The Difference Between Deep Copy and Shallow Copy

**Shallow Copy:**

A shallow copy creates a new object, but the contents (references to the objects in the original array) are not duplicated. Instead, references (pointers) to the original objects are copied.

If the original object contains mutable objects (like lists), modifications to those inner objects will affect both the original and the shallow copy.

```python
import copy
original_list = [[1, 2], [3, 4]]
shallow_copy = copy.copy(original_list)

shallow_copy[0][0] = 10
print(original_list)
```

Here, both original_list and shallow_copy point to the same inner lists. Changing one affects the other.

Deep Copy:

A deep copy creates a new object and recursively copies all objects within it. This means that the new object is completely independent of the original object.

Changes to mutable objects in the deep copy do not affect the original object.

```python
deep_copy = copy.deepcopy(original_list)

deep_copy[0][0] = 100
print(original_list)
print(deep_copy)
```

Here, original_list and deep_copy are fully independent. Modifying the deep copy does not affect the original.

Conclusion:

Shallow Copy: Copies the outer object, but inner objects (references) are shared between the original and the copy.

Deep Copy: Recursively copies both the outer object and all inner objects, creating an entirely independent copy.

**Question 14 : Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
Perform the following tasks using Matplotlib:**

**a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').**

**b)  Label the x-axis as 'Date'.**

**c) Set the title of the plot as 'Temperature and Humidity Over Time'.**


## Solution:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a date range
date_range = pd.date_range(start='2024-01-01', end='2024-01-10', freq='D')

# Step 2: Create a time-series dataset
data = {
    'Date': date_range,
    'Temperature': np.random.randint(15, 30, size=len(date_range)),  # Random temperatures between 15 and 30
    'Humidity': np.random.randint(40, 100, size=len(date_range))      # Random humidity between 40 and 100
}
df = pd.DataFrame(data)

# Step 3: Plot 'Temperature' and 'Humidity'
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], color='red', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Step 4: Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Step 5: Show the plot
plt.show()
```

**Question 15 : Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
tasks using Matplotlib:**

 **a) Plot a histogram of the data with 30 bins.**

 **b) Overlay a line plot representing the normal distribution's probability density function (PDF).**

 **c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.**

 **d) Set the title of the plot as 'Histogram with PDF Overlay'.**

## Solution:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Generate random samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)  # Mean = 0, Standard Deviation = 1

# Step 2: Plot a histogram
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Step 3: Overlay the PDF
# Calculate the PDF using the mean and standard deviation of the generated data
mean = np.mean(data)
std_dev = np.std(data)
xmin, xmax = plt.xlim()  # Get the range of the x-axis
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)  # Probability density function
plt.plot(x, p, 'k', linewidth=2)  # Overlay the PDF

# Step 4: Label the axes and set the title
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')

# Step 5: Show the plot
plt.grid()
plt.show()
```

**Question 17 : Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

## Solution:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate random arrays
np.random.seed(0)  # For reproducibility
x = np.random.rand(100) * 10  # Random values between 0 and 10
y = np.random.rand(100) * 10  # Random values between 0 and 10

# Step 2: Create a DataFrame to hold the data
data = pd.DataFrame({'X': x, 'Y': y})

# Step 3: Determine the quadrant for each point
def get_quadrant(row):
    if row['X'] >= 5 and row['Y'] >= 5:
        return 'Quadrant 1'
    elif row['X'] < 5 and row['Y'] >= 5:
        return 'Quadrant 2'
    elif row['X'] < 5 and row['Y'] < 5:
        return 'Quadrant 3'
    else:
        return 'Quadrant 4'

data['Quadrant'] = data.apply(get_quadrant, axis=1)

# Step 4: Create a scatter plot with Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='X', y='Y', hue='Quadrant', palette='Set1', style='Quadrant', s=100)

# Step 5: Customize the plot
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrants')
plt.axhline(5, color='grey', linewidth=0.8, linestyle='--')  # Horizontal line at y=5
plt.axvline(5, color='grey', linewidth=0.8, linestyle='--')  # Vertical line at x=5
plt.grid()

# Step 6: Show the plot
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Step 1: Prepare the data
x = np.linspace(0, 2 * np.pi, 100)  # x values from 0 to 2π
y = np.sin(x)                        # y values as sine of x

# Step 2: Create a new plot with title and axis labels
p = figure(title="Sine Wave Function",
           x_axis_label='X-axis (radians)',
           y_axis_label='Y-axis (sine values)',
           width=800, height=400)

# Step 3: Add a line renderer
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

# Step 4: Add grid lines
p.grid.grid_line_alpha = 0.5  # Adjust grid line transparency

# Step 5: Show the plot
output_notebook()  # Use this if you're in a Jupyter notebook
show(p)
```

**Question 19 : Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.**

## Solution:

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Step 1: Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']  # Define categories
values = np.random.randint(1, 100, size=len(categories))  # Random values for each category

# Step 2: Determine colors based on values
colors = ['#FF9999' if value < 50 else '#99FF99' for value in values]

# Step 3: Prepare data for Bokeh
source = ColumnDataSource(data=dict(categories=categories, values=values, colors=colors))

# Step 4: Create a new bar chart
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values',
           height=400, width=600)

# Step 5: Add bars to the chart
p.vbar(x='categories', top='values', source=source, width=0.9,
        color='colors', legend_field='categories')

# Step 6: Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Step 7: Customize the legend and grid
p.legend.title = "Categories"
p.legend.location = "top_left"
p.grid.grid_line_alpha = 0.3

# Step 8: Show the plot
output_notebook()  # Use this if you're in a Jupyter notebook
show(p)
```
